```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
# Saddle-node bifurcation
def saddle_node(x, y, a):
    u = a - x**2
    v = -y
    return u, v
```

```python
# Adjusted saddle-node bifurcation:
def adj_saddle_node(x, y, a):
    u = a - 2*(x**2) - 2
    v = -y
    return u, v
```

```python
# Phase portraits
# Width, height and points along axes
w = 2
h = 2
n = 100

# x and y dimensions
x_d = np.linspace(-w, w, n)
y_d = np.linspace(-h, h, n)

# Alpha values
a_values = np.array([-1., 0., 1., 5.])

# Meshgrid
x, y = np.meshgrid(x_d, y_d)

la = len(a_values)

# Figuresize
plt.figure(figsize=(10,5*la))

for i in range(la):
    a_val = a_values[i]
    
    splt = plt.subplot(la, 2, (2*i)+1)
    dx, dy = saddle_node(x, y, a_val)
    splt.streamplot(x, y, dx, dy, density=0.75)
    splt.set_title('alpha = ' + str(a_val))
    plt.xlabel('x1')
    plt.ylabel('x2')
    
    splt = plt.subplot(la, 2, (2*i)+2)
    dx, dy = adj_saddle_node(x, y, a_val)
    splt.streamplot(x, y, dx, dy, density=0.75)
    splt.set_title('alpha = ' + str(a_val))
    plt.xlabel('x1')
    plt.ylabel('x2')

plt.show()
```

```python
# Bifurcation diagrams
# Phase portraits
# Size and points along axes
s = 1
n = 100
xticks = np.linspace(-4, 4, 9)
yticks = np.linspace(-1, 1, 5)

# x_0 and alpha
x_0 = np.linspace(-s, s, n)
alpha_6 = x_0**2
alpha_7 = 2*x_0**2+2

# Figuresize
plt.figure(figsize=(10,5))


splt = plt.subplot(1,2,1)
splt.set_title('Bifurcation diagram Equation (6)')
splt.plot(alpha_6, x_0)
splt.set_xticks([-1.,-0.5, 0., 0.5,1])
splt.set_yticks(yticks)
plt.xlabel('alpha')
plt.ylabel('x_0s')

splt = plt.subplot(1,2,2)
splt.set_title('Bifurcation diagram Equation (7)')
splt.plot(alpha_7, x_0)
splt.set_xticks(xticks)
splt.set_yticks(yticks)
plt.xlabel('alpha')
plt.ylabel('x_0s')

plt.show()
```
